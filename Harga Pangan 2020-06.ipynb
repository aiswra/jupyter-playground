```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import plotly.graph_objects as go

%matplotlib inline
```

data source = https://hargapangan.id/tabel-harga/pasar-tradisional/daerah

```python
df = pd.read_csv("data/lh-hargapangan-jakarta.csv", skiprows=[0,1,2,3,4,5,6,7])
df.head(3)
```

```python
# Remove No column
df.drop(['No.'], axis=1, inplace=True)

# Rename colummn
df.rename(columns={"Komoditas(Rp)": "komoditas"}, inplace=True)

df.head(3)
```

```python
# Format value
date_cols = df.columns.to_list()
date_cols = date_cols[1:]

print(date_cols)
```

```python
for i in date_cols:
    df[i] = df[i].str.extract('(\d+,\d+)', expand=False).replace({',':''},regex=True).apply(pd.to_numeric,1)   
```

```python
df.head(3)
```

```python
# Add commodity name col
df['komoditas'].unique()
```

```python
list_komoditas = ['Beras', 'Daging Ayam','Daging Sapi','Telur Ayam',  'Bawang Merah','Bawang Putih','Cabai Merah', 'Cabai Rawit', 'Minyak Goreng',  'Gula Pasir']
len(list_komoditas)
```

```python
def getCommodityType(x):
    for i in list_komoditas:
        if i in x:
            return i
```

```python
df['tipe_komoditas'] = df['komoditas'].apply(lambda x: getCommodityType(x))
df.head(3)
```

```python
date_list = df.columns.to_list()
date_list.remove('komoditas')
date_list.remove('tipe_komoditas')
date_list
```

```python
dftemp =pd.melt(df, id_vars =['tipe_komoditas', 'komoditas'], value_vars = date_list) 
```

```python
dftemp['date'] = pd.to_datetime(dftemp['variable'], format = '%d/%m/%Y')
dftemp.head(3)
```

```python
# Plot - average

df_agg_type = dftemp.groupby(['tipe_komoditas','date'])['value'].mean().reset_index()

df_agg_type.head(2)
```

```python

list_tipe_komoditas = df_agg_type.tipe_komoditas.unique()
col_count = 1
row_count = 1
fig = make_subplots(
    rows=5, cols=2,
    subplot_titles=tuple(list_tipe_komoditas))

for i in list_tipe_komoditas:
    dft = df_agg_type.query('tipe_komoditas == "{}"'.format(i))
    fig.add_trace(go.Scatter(x=dft.date, y=dft.value,name = i),
              row=row_count, col=col_count)
    col_count +=1
    if col_count>2 : 
        col_count = 1
        row_count +=1
        if row_count>5 :
            row_count = 1


fig.update_layout(height=1400, width=1000,
                  title_text="Harga Pangan per Kg YTD")

fig.show()
```
